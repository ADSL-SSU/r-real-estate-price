```python
def string_to_list(keyword_str):
    result = []
    keyword_str = keyword_str.replace('[','')
    keyword_str = keyword_str.replace(']','')
    keyword_str = keyword_str.replace('(','')
    keyword_str = keyword_str.replace(')','')
    keyword_str = keyword_str.replace("'",'')
    keyword_str = keyword_str.replace('""','')
    keyword_list = keyword_str.split(',')
    for i in range(0,len(keyword_list),2):
        result.append(keyword_list[i:i+2])
    return result
```

```python
import pandas as pd

read_path = '../article/headline_keyword/'
write_path = '../article/headline_keyword/'
names = ['date','week','headline','keyword']
all_article = pd.DataFrame({
    'date' : [],
    'headline': [],
    'keyword' : []
})

for week in range(783):
    df = pd.read_csv(f'{read_path}week{week}_headline_keyword.csv')
    df['week'] = week
    all_article = pd.concat([all_article, df], ignore_index=1)
all_article['week'] = all_article['week'].astype(int)
all_article['keyword'] = [string_to_list(keyword) for keyword in all_article['keyword']]
```

```python
key_word_kind = []
for week in range(783):
    keyWord_dict = {}
    for word_list in all_article['keyword'][all_article['week'] == week]:
        for word in word_list:
            if word[1] not in key_word_kind:
                key_word_kind.append(word[1])
key_word_kind
```

```python
def word_sorting(words):
    table = []
    for week in range(783):
        keyWord_dict = {}
        for word_list in all_article['keyword'][all_article['week'] == week]:
            for word in word_list:
                word[0] = word[0].replace(' ','')
                if word[1] not in words: continue
                if word[0] in keyWord_dict:
                    keyWord_dict[word[0]] += 1
                else:
                    keyWord_dict[word[0]] = 1
        if len(keyWord_dict) <= 0: continue
        s = pd.Series(keyWord_dict.values(),keyWord_dict.keys())
        table.append(s)
    return table
```

```python
def word_columns(words):
    table = []
    for week in range(783):
        keyWord_dict = {}
        keyWord_dict['total'] = 0
        for word_list in all_article['keyword'][all_article['week'] == week]:
            for word in word_list:
                word[0] = word[0].replace(' ','')
                if word[0] not in words: continue
                if word[0] in keyWord_dict:
                    keyWord_dict[word[0]] += 1
                else:
                    keyWord_dict[word[0]] = 1
                keyWord_dict['total'] += 1
        if len(keyWord_dict) <= 0: continue
        s = pd.Series(keyWord_dict.values(),keyWord_dict.keys())
        table.append(s)
    return table
```

```python
keyword_table = pd.DataFrame(word_sorting([' CITY',' LOCATION']))
```

```python
keyword_table = keyword_table.fillna(0)
keyword_table.to_csv(f'{write_path}keyword_about_location.csv')
```
