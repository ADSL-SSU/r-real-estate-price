```python
from pororo import Pororo
ner = Pororo(task="ner", lang="ko")
```

```python
import pandas as pd

read_path = '../article/monthly_article/'
write_path = '../article/headline_keyword/'
#column_name = ['date','cat','co','headline','text','url']
article = []

for year in range(2006,2021):
    for month in range(1,13):
        date = year*100 + month
        article.append(pd.read_csv(f'{read_path}Article_부동산_{date}_{date}.csv', names = ['date','cat','com','headline','text','url'], encoding = 'cp949'))
```

```python
article[0]
```

```python
i = 0
for year in range(2006,2021):
    for month in range(1,13):
        date = year*100 + month
        print(f'>>> start {date} article headline keyword crawling')
        df = pd.DataFrame({
            'date' : article[i]['date'],
            'headline' : article[i]['headline'],
            'keyword' : [ner(headline) for headline in article[i]['headline']]
        })
        df.to_csv(f'{write_path}{date}_headline_keyword.csv')
        print(f'>>> complete {date} article headline keyword crawling')
        i += 1
```
