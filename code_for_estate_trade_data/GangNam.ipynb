```python
# 서울 아파트 매매 데이터 호출
import pandas as pd

trade_path = '../seoul_apt_trade_data/'
rent_path = '../seoul_apt_rent_data/'

trade_gangnam = pd.read_csv(f'{trade_path}강남구_trade.csv', encoding='CP949')
rent_gangnam = pd.read_csv(f'{rent_path}강남구_rent.csv', encoding='CP949')
```

```python
# 거래일 생성기
def make_date(year, month):
    date = ''
    if month > 12:
        date = str(year+1) + '-' + '01' + '-00'
    elif month >= 10:
        date = str(year) + '-' + str(month) + '-00'
    else: 
        date = str(year) + '-0' + str(month) + '-00'
    return date
```

```python
count_trade_gangnam =  [trade_gangnam[(trade_gangnam.iloc[:,3] > make_date(year,month)) 
                                 & (trade_gangnam.iloc[:,3] < make_date(year,month+1))].iloc[:,0].count()
                               for year in range(2006,2021) for month in range(1,13)]
```

```python
# 강남 아파트 매매 월간 통계
import matplotlib.pyplot as plt

plt.plot(count_trade_gangnam)
```

```python
count_rent_gangnam =  [rent_gangnam[(rent_gangnam.iloc[:,3] > make_date(year,month)) 
                                 & (rent_gangnam.iloc[:,3] < make_date(year,month+1))].iloc[:,0].count()
                               for year in range(2011,2021) for month in range(1,13)]
```

```python
# 강남 아파트 전세 월간 통계
plt.plot(count_rent_gangnam)
```
